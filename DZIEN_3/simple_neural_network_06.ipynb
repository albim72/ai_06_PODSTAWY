```python
import tensorflow as tf
```

```python
mnist = tf.keras.datasets.mnist
```

```python
(x_train,y_train),(x_test,y_test) = mnist.load_data()
```

```python
x_train, x_test = x_train/255.0, x_test/255.0
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#budowa sieci neuronowej
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(128,activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(10)
])

model.summary()
```

```python
model.compile(optimizer="adam",
              loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics = ['accuracy']
              )
```

```python
model.fit(x_train,y_train,epochs=10)
```

```python
#ocena modelu
test_loss, test_acc = model.evaluate(x_test,y_test)
print(f"strata: {test_loss}")
print(f"accuracy: {test_acc}")
```

```python
#predykcja
logits = model.predict(x_test)

print(f"kształt logit: {logits.shape}")
```

```python
#zamian logitów na prawdopodobięnstwa

import numpy as np
y_proba = tf.nn.softmax(logits,axis=1).numpy()

print(f"kształt prawdopodobieństw : {y_proba.shape}")
print(f"przypadek 0 -> {y_proba[0]}")
print(f"przypadek 0 -> {np.argmax(y_proba[0])}")
print(f"test -> {y_test[0]}")

print(f"przypadek 121 -> {y_proba[121]}")
print(f"przypadek 121 -> {np.argmax(y_proba[121])}")
print(f"test -> {y_test[121]}")
```

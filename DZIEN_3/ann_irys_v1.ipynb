```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_iris
```

```python
# Wczytanie zbioru danych Iris
iris = load_iris()

# Dane wejściowe
X = iris.data

# Etykiety klas
y = iris.target

# Nazwy cech
feature_names = iris.feature_names

# Nazwy klas
target_names = iris.target_names

print("Podstawowe informacje o zbiorze Iris")
print("------------------------------------")

print("Liczba próbek:", X.shape[0])
print("Liczba cech:", X.shape[1])
print("Kształt X:", X.shape)
print("Kształt y:", y.shape)

print("\nNazwy cech:")
print(feature_names)

print("\nDostępne klasy:")
print(target_names)

print("\nPierwsze 5 próbek danych wejściowych:")
print(X[:5])

print("\nPierwsze 5 etykiet:")
print(y[:5])
```

```python
# Zamiana danych na DataFrame

df = pd.DataFrame(X, columns=feature_names)

# Dodanie kolumny z etykietą liczbową
df["target"] = y

# Dodanie kolumny z nazwą gatunku
df["species"] = df["target"].map({
    0: target_names[0],
    1: target_names[1],
    2: target_names[2]
})

df.head()
```

```python
# Informacje o danych

print("Informacje o DataFrame:")
print(df.info())

print("\nPodstawowe statystyki:")
print(df.describe())

print("\nLiczba przykładów w każdej klasie:")
print(df["species"].value_counts())
```
